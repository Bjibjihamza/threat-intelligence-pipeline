```python
import re
import math
import pandas as pd
from collections import Counter, defaultdict

# =========================================
# CONFIG
# =========================================
CSV_PATH = "stacked_deduplicated_dataset.csv"
OUT_PATH = "raw_cve_details_with_predicted_category.csv"

ALPHA = 1.0          # Laplace smoothing
MIN_TOKEN_LEN = 2
USE_BIGRAMS = True

STOPWORDS = {
    "the","a","an","and","or","of","to","in","for","on","with","by","via",
    "from","at","as","is","are","be","was","were","this","that","these","those",
    "vulnerability","vulnerabilities","issue","bug","attack","remote","local",
    "privilege","escalation","denial","service","dos","code","execution","rce",
    "information","disclosure","leak","overflow","buffer","command","injection",
    "authentication","bypass","default","config","configuration","file","read",
    "write","arbitrary","sun","sunos","solaris","unix","linux","hp","ibm","sgi",
    "irix","cisco","apache","server","client","kernel"
}

# =========================================
# TOKENIZATION
# =========================================
non_alnum = re.compile(r"[^a-z0-9]+")

def tokenize(text: str):
    text = str(text or "").lower()
    text = non_alnum.sub(" ", text)
    toks = [t for t in text.split() if len(t) >= MIN_TOKEN_LEN and t not in STOPWORDS]
    if USE_BIGRAMS and len(toks) >= 2:
        bigrams = [f"{toks[i]}_{toks[i+1]}" for i in range(len(toks)-1)]
        toks.extend(bigrams)
    return toks

# =========================================
# TRAIN MODEL FROM EXISTING CATEGORY
# =========================================
def train_keyword_model(df: pd.DataFrame):
    labeled = df[df["category"].notna() & (df["category"].astype(str).str.strip() != "")]
    if labeled.empty:
        raise ValueError("⚠️ No labeled rows found in 'category'.")

    cat_doc_counts = Counter()
    cat_token_counts = defaultdict(Counter)
    vocab = Counter()

    for _, row in labeled.iterrows():
        cat = str(row["category"])
        cat_doc_counts[cat] += 1
        tokens = tokenize(row.get("title", ""))
        cat_token_counts[cat].update(tokens)
        vocab.update(tokens)

    total_docs = sum(cat_doc_counts.values())
    priors = {cat: math.log(count / total_docs) for cat, count in cat_doc_counts.items()}
    total_tokens = {cat: sum(cnt.values()) for cat, cnt in cat_token_counts.items()}
    vocab_size = len(vocab)

    return {
        "priors_log": priors,
        "token_counts": cat_token_counts,
        "total_tokens": total_tokens,
        "vocab_size": vocab_size,
        "alpha": ALPHA
    }

# =========================================
# PREDICT FUNCTION
# =========================================
def predict_category(title: str, model) -> str:
    tokens = tokenize(title)
    if not tokens:
        return "Unknown"

    priors_log = model["priors_log"]
    token_counts = model["token_counts"]
    total_tokens = model["total_tokens"]
    V = model["vocab_size"]
    alpha = model["alpha"]

    best_cat, best_score = None, -float("inf")
    for cat in priors_log.keys():
        score = priors_log[cat]
        denom = total_tokens[cat] + alpha * V
        counts = token_counts[cat]
        for tok in tokens:
            score += math.log((counts.get(tok, 0) + alpha) / denom)
        if score > best_score:
            best_score, best_cat = score, cat

    return best_cat if best_cat else "Unknown"

# =========================================
# MAIN PREDICTION CYCLE
# =========================================
def run_prediction_cycle():
    df = pd.read_csv(CSV_PATH, low_memory=False)

    # Create the column if missing
    if "Predicted_category" not in df.columns:
        df["Predicted_category"] = ""

    # Train model
    model = train_keyword_model(df)

    # Predict for ALL rows (including empty or NaN Predicted_category)
    df["Predicted_category"] = df["title"].apply(lambda t: predict_category(t, model))

    # If a true category exists, keep it (you can remove this line if you want model overrides)
    #has_true_cat = df["category"].notna() & (df["category"].astype(str).str.strip() != "")
    #df.loc[has_true_cat, "Predicted_category"] = df.loc[has_true_cat, "category"].astype(str)

    # Replace any remaining NaN/empty with "Unknown"
    df["Predicted_category"] = df["Predicted_category"].fillna("Unknown").replace(r"^\s*$", "Unknown", regex=True)

    # Save final dataset
    df.to_csv(OUT_PATH, index=False)
    print(f"✅ Predictions completed and saved to -> {OUT_PATH}")
    print(df[["title", "category", "Predicted_category"]].head(5))
    return df

# =========================================
# EXECUTE
# =========================================
if __name__ == "__main__":
    run_prediction_cycle()
```

```python
import pandas as pd
import os

# Step 2: Iterate through the unique categories (excluding 'Unknown')
categories = df4['category'].unique()

# Step 3: Create a directory to store the CSV files (optional)
output_dir = 'extracted_description'
os.makedirs(output_dir, exist_ok=True)

# Step 4: Loop through each category, filter the dataset, and save titles to CSV
for category in categories:
    # Skip 'Unknown' category
    if category == 'Unknown':
        continue
    
    # Ensure category is a string (in case it's a number or NaN after filling)
    category = str(category)
    
    # Filter rows belonging to the current category
    category_df = df4[df4['category'] == category]
    
    # Create a filename from the category (replace spaces with underscores)
    filename = os.path.join(output_dir, f"{category.replace(' ', '_')}.csv")
    
    # Save the titles to the CSV file
    category_df[['cve_id', 'description']].to_csv(filename, index=False)
    print(f"Saved descriptions for category '{category}' to {filename}")

print("\nDone! Unknown category was skipped.")
```

```python
import pandas as pd
import os
from collections import Counter
import re
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
import time

start_time = time.time()

# Step 1: Read all CSV files from extracted_description folder
extracted_dir = 'extracted_description'
category_data = {}

print("Extracting data from each category...\n")

for filename in os.listdir(extracted_dir):
    if filename.endswith('.csv') and filename != 'Unknown.csv':
        category = filename.replace('.csv', '').replace('_', ' ')
        filepath = os.path.join(extracted_dir, filename)
        df = pd.read_csv(filepath)
        descriptions = df['description'].astype(str).tolist()
        category_data[category] = descriptions
        print(f"{category}: {len(descriptions)} samples")

print("\n" + "="*80)
print("Building TF-IDF model...\n")

# Step 2: Build TF-IDF vectors
all_category_texts = []
category_labels = []

for category, descriptions in category_data.items():
    combined_text = ' '.join(descriptions)
    all_category_texts.append(combined_text)
    category_labels.append(category)

tfidf = TfidfVectorizer(
    max_features=1000,
    stop_words='english',
    ngram_range=(1, 3),
    min_df=2,
    max_df=0.7
)

category_vectors = tfidf.fit_transform(all_category_texts)

print(f"TF-IDF model built with {len(tfidf.vocabulary_)} features")
print(f"Time elapsed: {time.time() - start_time:.2f} seconds\n")

# Step 3: BATCH PREDICTION (Much faster!)
print("="*80)
print(f"Predicting categories for {len(df4)} rows using batch processing...\n")

# Fill NaN descriptions with empty string
df4['description'] = df4['description'].fillna('')

# Transform all descriptions at once (MUCH faster than one-by-one)
all_desc_vectors = tfidf.transform(df4['description'].astype(str))

# Calculate similarities for all descriptions at once
all_similarities = cosine_similarity(all_desc_vectors, category_vectors)

# Get the category with highest similarity for each row
predictions = [category_labels[np.argmax(sim)] for sim in all_similarities]

# Add the new column
df4['predicted_cat_desc'] = predictions

print(f"Prediction complete!")
print(f"Total time: {time.time() - start_time:.2f} seconds\n")

print(f"Category distribution:")
print(df4['predicted_cat_desc'].value_counts())

# Calculate accuracy if category column exists
if 'category' in df4.columns:
    known_mask = df4['category'] != 'Unknown'
    if known_mask.sum() > 0:
        known_rows = df4[known_mask]
        correct = (known_rows['category'] == known_rows['predicted_cat_desc']).sum()
        total = len(known_rows)
        accuracy = (correct / total) * 100
        
        print(f"\n{'='*80}")
        print(f"ACCURACY ON KNOWN CATEGORIES")
        print(f"{'='*80}")
        print(f"Accuracy: {accuracy:.2f}%")
        print(f"Correct: {correct} out of {total}")

# Save results
output_file = 'df4_with_desc_predictions.csv'
df4.to_csv(output_file, index=False)
print(f"\nUpdated dataframe saved to: {output_file}")

print(f"\nTotal execution time: {time.time() - start_time:.2f} seconds")
print("Done!")
```

```python
import pandas as pd

print("="*80)
print("ACCURACY COMPARISON")
print("="*80)

# Filter rows with known categories (not Unknown or NaN)
known_mask = (df5['category'].notna()) & (df5['category'] != 'Unknown')
known_rows = df5[known_mask]

print(f"\nTotal rows with known categories: {len(known_rows)}")
print(f"Total rows in dataset: {len(df5)}")

# ============= ACCURACY 1: Category vs Predicted Category (from titles) =============
print(f"\n{'='*80}")
print("1. ACCURACY: Category vs Predicted Category (from titles)")
print(f"{'='*80}")

if 'Predicted_category' in df5.columns:
    correct_title = (known_rows['category'] == known_rows['Predicted_category']).sum()
    total = len(known_rows)
    accuracy_title = (correct_title / total) * 100
    
    print(f"Accuracy: {accuracy_title:.2f}%")
    print(f"Correct: {correct_title} out of {total}")
    print(f"Incorrect: {total - correct_title}")
else:
    print("Column 'Predicted_category' not found!")
    accuracy_title = None

# ============= ACCURACY 2: Category vs Predicted_cat_desc (from descriptions) =============
print(f"\n{'='*80}")
print("2. ACCURACY: Category vs Predicted_cat_desc (from descriptions)")
print(f"{'='*80}")

if 'predicted_cat_desc' in df5.columns:
    correct_desc = (known_rows['category'] == known_rows['predicted_cat_desc']).sum()
    total = len(known_rows)
    accuracy_desc = (correct_desc / total) * 100
    
    print(f"Accuracy: {accuracy_desc:.2f}%")
    print(f"Correct: {correct_desc} out of {total}")
    print(f"Incorrect: {total - correct_desc}")
else:
    print("Column 'predicted_cat_desc' not found!")
    accuracy_desc = None

# ============= COMPARISON =============
print(f"\n{'='*80}")
print("COMPARISON SUMMARY")
print(f"{'='*80}")

if accuracy_title is not None and accuracy_desc is not None:
    print(f"\nPrediction from Titles:       {accuracy_title:.2f}%")
    print(f"Prediction from Descriptions: {accuracy_desc:.2f}%")
    
    difference = accuracy_desc - accuracy_title
    if difference > 0:
        print(f"\n✓ Descriptions are MORE accurate by {abs(difference):.2f}%")
    elif difference < 0:
        print(f"\n✓ Titles are MORE accurate by {abs(difference):.2f}%")
    else:
        print(f"\n✓ Both methods have the SAME accuracy")
    
    # Show which method is better
    if accuracy_desc > accuracy_title:
        print(f"   → Use 'predicted_cat_desc' for better results")
    elif accuracy_title > accuracy_desc:
        print(f"   → Use 'predicted_category' for better results")

print("\n" + "="*80)
```
